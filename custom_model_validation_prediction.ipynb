###
Prediction of the three models

```python
import subprocess
import os
import time
# Directory containing the images
image_dir = "./test_dataset_yolo/"

# List all files in the directory
image_files = [f for f in os.listdir(image_dir) if os.path.isfile(os.path.join(image_dir, f))]

# Iterate over each image file
count =0
for image_file in image_files:
    # Construct the YOLOv8 inference command
    command = f"yolo task=detect mode=predict model=runs/detect/yolov8n_e509/weights/best.pt source={os.path.join(image_dir, image_file)} show=True imgsz=640 name=yolov8n_v8_50e_infer640"
    # Execute the command using the subprocess module
    subprocess.run(command, shell=True)
    # test yolo small model
    command = f"yolo task=detect mode=predict model=runs/detect/yolov8s_e50/weights/best.pt source={os.path.join(image_dir, image_file)} show=True imgsz=640 name=yolov8s_v8_50e_infer640"
    subprocess.run(command, shell=True)
    # test yolo meduim model
    command = f"yolo task=detect mode=predict model=runs/detect/yolov8m_e5/weights/best.pt source={os.path.join(image_dir, image_file)} show=True imgsz=640 name=yolov8m_v8_50e_infer640"
    subprocess.run(command, shell=True)
    count+=1
    if count ==5:
        break
```
